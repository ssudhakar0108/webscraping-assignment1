# Write a python program to scrape ‘software developer’ job listings from ‘Monster.com’. It should include allthe jobs listed for the next 5 pages

```python
import requests
from bs4 import BeautifulSoup as soup

```

```python
import requests
URL='https://www.monster.com/jobs/search/?q=Software-Developer&where=india&stpage=1&page=2'
```

```python
page = requests.get(URL)
```

```python
bsobj=soup(page.content,'lxml')
```

```python

title = []
for header in bsobj.findAll('h2',{'class':'title'}):
  title.append(header.text.strip())
```

```python
title
```

```python
company = []
for name in bsobj.findAll('div',{'class':'company'}):
  company.append(name.text.strip())
```

```python
company
```

```python
location = []
for locs in bsobj.findAll('div',{'class':'location'}):
  location.append(locs.text.strip())
location = location[1:]
```

```python
location
```

```python

import pandas as pd
d1 = {'title':title,'company':company,'location':location}
SoftwareDeveloper_Monster= pd.DataFrame.from_dict(d1)
```

```python
SoftwareDeveloper_Monster
```

```python
import csv
#save to CSV
SoftwareDeveloper_Monster.to_csv('SoftwareDeveloper_Monster.csv')
```
