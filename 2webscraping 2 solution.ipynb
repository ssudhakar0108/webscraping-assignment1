```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver=webdriver.Chrome(r"/home/sudhakar/Downloads/chromedriver")
```

```python
url="https://www.imdb.com/chart/top"
```

```python
driver.get(url)
```

```python
Name=[]
IMBD_ratings=[]
Year_of_release=[]
```

```python
name_tags=driver.find_elements_by_xpath("//td[@class='titleColumn']")
```

```python
name_tags
```

```python
for i in name_tags:
    name1=i.text
    Name.append(name1)
 
```

```python
Name
```

```python
imbd_rating_tags=driver.find_elements_by_xpath("//td[@class='ratingColumn imdbRating']")
```

```python
imbd_rating_tags
```

```python
for i in imbd_rating_tags:
    rating=i.text
    IMBD_ratings.append(rating)
```

```python
IMBD_ratings
```

```python
year_of_release_tags=driver.find_elements_by_xpath("//span[@class='secondaryInfo']")
```

```python
year_of_release_tags
```

```python
for i in year_of_release_tags:
    release1=i.text
    Year_of_release.append(release1)
    
    
```

```python
Year_of_release
```

```python
print(len(Name),len(IMBD_ratings),len(Year_of_release))
```

```python
import pandas as pd
movie=pd.DataFrame({})
movie['Name']=Name
movie['IMBD_ratings']=IMBD_ratings
movie['Year_of_release']=Year_of_release
```

```python
movie
```

```python
import csv
#save to CSV
movie.to_csv('movie.csv')

 
```
